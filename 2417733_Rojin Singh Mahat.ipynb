```python
#Task - 3 - Outlier Detection:
#1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
#2. Define outliers using the 1.5 Ã— IQR rule.
#3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
#in a different color.
#4. Discuss the characteristics of these outliers and their potential impact on regional averages.
# Function to calculate outliers using the 1.5 * IQR rule 
def find_outliers(data, column): 
    Q1 = data[column].quantile(0.25) 
    Q3 = data[column].quantile(0.75) 
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR 
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)] 
    return outliers
```

```python
# Identify outliers for GDP per Capita
gdp_outliers = find_outliers(southAsia_df, 'Log GDP per capita')

# Identify outliers for Score
score_outliers = find_outliers(southAsia_df, 'score')

# Combine outliers
outliers = pd.concat([gdp_outliers, score_outliers]).drop_duplicates()

# Display the outliers
print("Outliers based on GDP per Capita and Score:")
outliers[['Country name', 'Log GDP per capita', 'score']]
```

```python
# Create a scatter plot highlighting outliers
plt.figure(figsize=(10, 6))
plt.scatter(southAsia_df['Log GDP per capita'], southAsia_df['score'], label='Normal')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='r', label='Outlier')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot of GDP per Capita vs Score with Outliers Highlighted')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the aquired scatter plot, we can observe that outliers like afghanistan and sri lanka can put noticable impact on the regional averages. Afghanistan with it's low score can bring down the regional averages while Sri Lanka with its high scores can bring the regional averaes higher, revealing stark inequalities and need for interventions for certain datas. Understanding these outliers helps in crafting more effective policies and addressing the underlying issues specific to these regions.
